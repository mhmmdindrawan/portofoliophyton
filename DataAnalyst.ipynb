```python
import seaborn as sns

# Create a count plot for 'Gender' and 'MaritalStatus'
sns.countplot(data=df_personaldata, x='Gender', hue='MaritalStatus')

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Marital Status by Gender')

# Show the plot
plt.show()
```
